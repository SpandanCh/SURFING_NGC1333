```python
for i in range(tp_master_13CO_3sigma.shape[1]):
    for j in range(tp_master_13CO_3sigma.shape[2]):
#         sort_para_all = tp_master_13CO_3sigma[:, i, j].copy()
        sort_para_all = sig_master_13CO_3sigma[:, i, j].copy()
#         sort_para_all = mom0_master_13CO_3sigma[:, i, j].copy()
        
        if np.isfinite(np.nanmax(sort_para_all)):
            
            loc_1 = np.where(sort_para_all == np.nanmax(sort_para_all))[0][0]
            dummy_arr_1[0, i, j] = tp_master_13CO_3sigma[loc_1, i, j]
            dummy_arr_1[1, i, j] = vel_master_13CO_3sigma[loc_1, i, j]
            dummy_arr_1[2, i, j] = sig_master_13CO_3sigma[loc_1, i, j]
            dummy_arr_1[3, i, j] = mom0_master_13CO_3sigma[loc_1, i, j]
            sort_para_all[loc_1] = np.nan
            
            if np.isfinite(np.nanmax(sort_para_all)):
                loc_2 = np.where(sort_para_all == np.nanmax(sort_para_all))[0][0]
                dummy_arr_2[0, i, j] = tp_master_13CO_3sigma[loc_2, i, j]
                dummy_arr_2[1, i, j] = vel_master_13CO_3sigma[loc_2, i, j]
                dummy_arr_2[2, i, j] = sig_master_13CO_3sigma[loc_2, i, j]
                dummy_arr_2[3, i, j] = mom0_master_13CO_3sigma[loc_2, i, j]
                sort_para_all[loc_2] = np.nan
                
                if np.isfinite(np.nanmax(sort_para_all)):
                    loc_3 = np.where(sort_para_all == np.nanmax(sort_para_all))[0][0]
                    dummy_arr_3[0, i, j] = tp_master_13CO_3sigma[loc_3, i, j]
                    dummy_arr_3[1, i, j] = vel_master_13CO_3sigma[loc_3, i, j]
                    dummy_arr_3[2, i, j] = sig_master_13CO_3sigma[loc_3, i, j]
                    dummy_arr_3[3, i, j] = mom0_master_13CO_3sigma[loc_3, i, j]
                    sort_para_all[loc_3] = np.nan
                    
                    if np.isfinite(np.nanmax(sort_para_all)):
                        loc_4 = np.where(sort_para_all == np.nanmax(sort_para_all))[0][0]
                        dummy_arr_4[0, i, j] = tp_master_13CO_3sigma[loc_4, i, j]
                        dummy_arr_4[1, i, j] = vel_master_13CO_3sigma[loc_4, i, j]
                        dummy_arr_4[2, i, j] = sig_master_13CO_3sigma[loc_4, i, j]
                        dummy_arr_4[3, i, j] = mom0_master_13CO_3sigma[loc_4, i, j]
                        sort_para_all[loc_4] = np.nan
                        
                        if np.isfinite(np.nanmax(sort_para_all)):
                            loc_5 = np.where(sort_para_all == np.nanmax(sort_para_all))[0][0]
                            dummy_arr_5[0, i, j] = tp_master_13CO_3sigma[loc_5, i, j]
                            dummy_arr_5[1, i, j] = vel_master_13CO_3sigma[loc_5, i, j]
                            dummy_arr_5[2, i, j] = sig_master_13CO_3sigma[loc_5, i, j]
                            dummy_arr_5[3, i, j] = mom0_master_13CO_3sigma[loc_5, i, j]
                            sort_para_all[loc_5] = np.nan

                            if np.isfinite(np.nanmax(sort_para_all)):
                                loc_6 = np.where(sort_para_all == np.nanmax(sort_para_all))[0][0]
                                dummy_arr_6[0, i, j] = tp_master_13CO_3sigma[loc_6, i, j]
                                dummy_arr_6[1, i, j] = vel_master_13CO_3sigma[loc_6, i, j]
                                dummy_arr_6[2, i, j] = sig_master_13CO_3sigma[loc_6, i, j]
                                dummy_arr_6[3, i, j] = mom0_master_13CO_3sigma[loc_6, i, j]
                                sort_para_all[loc_6] = np.nan
                                
                                if np.isfinite(np.nanmax(sort_para_all)):
                                    loc_7 = np.where(sort_para_all == np.nanmax(sort_para_all))[0][0]
                                    dummy_arr_7[0, i, j] = tp_master_13CO_3sigma[loc_7, i, j]
                                    dummy_arr_7[1, i, j] = vel_master_13CO_3sigma[loc_7, i, j]
                                    dummy_arr_7[2, i, j] = sig_master_13CO_3sigma[loc_7, i, j]
                                    dummy_arr_7[3, i, j] = mom0_master_13CO_3sigma[loc_7, i, j]
                                    sort_para_all[loc_7] = np.nan

                                    if np.isfinite(np.nanmax(sort_para_all)):
                                        loc_8 = np.where(sort_para_all == np.nanmax(sort_para_all))[0][0]
                                        dummy_arr_8[0, i, j] = tp_master_13CO_3sigma[loc_8, i, j]
                                        dummy_arr_8[1, i, j] = vel_master_13CO_3sigma[loc_8, i, j]
                                        dummy_arr_8[2, i, j] = sig_master_13CO_3sigma[loc_8, i, j]
                                        dummy_arr_8[3, i, j] = mom0_master_13CO_3sigma[loc_8, i, j]
                                        sort_para_all[loc_8] = np.nan

                                        if np.isfinite(np.nanmax(sort_para_all)):
                                            loc_9 = np.where(sort_para_all == np.nanmax(sort_para_all))[0][0]
                                            dummy_arr_9[0, i, j] = tp_master_13CO_3sigma[loc_9, i, j]
                                            dummy_arr_9[1, i, j] = vel_master_13CO_3sigma[loc_9, i, j]
                                            dummy_arr_9[2, i, j] = sig_master_13CO_3sigma[loc_9, i, j]
                                            dummy_arr_9[3, i, j] = mom0_master_13CO_3sigma[loc_9, i, j]
                                            sort_para_all[loc_9] = np.nan

        rem_arr[:, i, j] = sort_para_all
```
